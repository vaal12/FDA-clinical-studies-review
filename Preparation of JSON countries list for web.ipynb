```python
import pandas as pd
import numpy as np
```

```python
countries_full_df = pd.read_csv(
    r"c:\Dev\04. Python\03. XML converter of FDA list\2018Oct30_00-15-03\FDA_DB_COUNTRY_LOC_TABLE_2018Oct30_00-15-03.csv")

#Data cleaning
countries_full_df['country'][countries_full_df['country'] == "United States"] = "United States of America"
countries_full_df['country'][countries_full_df['country'] == "Congo, The Democratic Republic of the"] = "Congo"
countries_full_df['country'][countries_full_df['country'] == "Czech Republic"] = "Czechia"
```

```python
counted_countries = countries_full_df.groupby('country').count()

sorted_counted_countries = pd.DataFrame(counted_countries.sort_values(by="nct_id", ascending=False))
sorted_counted_countries.head()
```

```python
countries_codes = pd.read_csv(
    r"c:\Dev\05. Jupyter\2. FDA countries clean\csv\countries_codes_3Nov.csv"
)
countries_codes.head()
```

```python
coded_countries = sorted_counted_countries.merge(countries_codes, 
                            how="left", left_on="country", right_on="full_country_name")
coded_countries.drop(["Unnamed: 0"], axis=1, inplace=True)
coded_countries[coded_countries.full_country_name.isnull()].head()
```

```python
coded_countries.rename(columns={"nct_id":"nct_id_count"}, inplace=True)
coded_countries = coded_countries.set_index("iso2")
coded_countries.head()
```
