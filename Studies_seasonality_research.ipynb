```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# %matplotlib notebook
%matplotlib inline
```

```python
studies_full_df = pd.read_csv(
    r"c:\Dev\04. Python\03. XML converter of FDA list\goodDB\05. 2018Sep22_11-50-07\FDA_DB_MAIN_TABLE_2018Sep22_11-50-07.csv",
    parse_dates = ["study_first_submitted"]
)
studies_full_df.head()
```

```python
studies_full_df.info()
```

```python
culled_studies_full_info_df = pd.DataFrame(studies_full_df[::1000])
len(culled_studies_full_info_df)
```

```python
culled_studies_full_info_df.head()
```

```python
# culled_studies_full_info_df["study_first_submitted"].dt.month
culled_studies_full_info_df = culled_studies_full_info_df.assign(
    study_first_submitted_month = np.int64(culled_studies_full_info_df["study_first_submitted"].dt.month))
culled_studies_full_info_df = culled_studies_full_info_df.assign(
    study_first_submitted_year = np.int64(culled_studies_full_info_df["study_first_submitted"].dt.year))

# culled_studies_full_info_df["study_first_submitted_month"] = culled_studies_full_info_df["study_first_submitted"].values.astype('datetime64[M]')
culled_studies_full_info_df.head()
```

```python
studies_by_month = pd.DataFrame(culled_studies_full_info_df.groupby("study_first_submitted_month").count())
studies_by_month.drop(columns=[
    "org_study_id", 
    "study_first_submitted_year", 
    "url", 
    "brief_title", 
    "official_title", 
    "source", 
    "brief_summary", 
    "detailed_description", 
    "overall_status", 
    "last_known_status", 
    "start_date", 
    "completion_date", 
    "phase", 
    "study_type", 
    "has_expanded_access", 
    "allocation", 
    "intervention_model", 
    "primary_purpose", 
    "masking", 
    "study_first_submitted", 
    "study_first_submitted_qc", 
    "last_update_submitted", 
    "qwe1"
    ], inplace=True)
studies_by_month.head()
```
