# Country population table clean and merge

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# %matplotlib notebook
%matplotlib inline
```

```python
countries_full_df = pd.read_csv(
    r"c:\Dev\04. Python\03. XML converter of FDA list\goodDB\04. 2018Sep17_13-28-24\FDA_DB_COUNTRY_LOC_TABLE_2018Sep17_13-28-24.csv")
counted_countries = countries_full_df.groupby('country').count()
counted_countries.head()
```

```python
#Cleaning of counted_countries
counted_countries_clean = pd.DataFrame(counted_countries[counted_countries.index != "**None**"])
counted_countries_clean["country_lower"] = counted_countries_clean.index.str.lower()
counted_countries_clean.head()
```

```python
countries_population_df = pd.read_csv(
    r"c:\Dev\05. Jupyter\2. FDA countries clean\Countries_population_2018.csv",
#     dtype = {"population_fig":np.int64}
)
# countries_population_df.head()
countries_population_df["parsed_population_fig"] = np.int64(countries_population_df["population_fig"].str.replace(',', ''))
countries_population_df["country_lower"] = countries_population_df["country"].str.lower()
countries_population_df.head()
```

```python
countries_population_df.info()
```

```python
merged_country_populace_df = pd.merge(
    counted_countries_clean, 
    countries_population_df, 
    left_on = "country_lower",
    right_on = "country_lower",
    how='left'
)
merged_country_populace_df.head()
```

```python
merged_country_populace_df.sort_values("parsed_population_fig").head()
```

```python
nan_countries = pd.DataFrame(merged_country_populace_df.loc[
    pd.isnull(merged_country_populace_df.parsed_population_fig)])
nan_countries.shape
```
